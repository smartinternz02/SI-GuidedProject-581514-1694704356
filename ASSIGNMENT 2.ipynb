```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Loading the dataset

```python
df=pd.read_csv("House Price India.csv")
```

```python
df.head()
```

# Univariate Analysis

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['number of bedrooms'], bins=20)
plt.xlabel('number of bedrooms')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
bedroom=df['number of bedrooms'].value_counts().sort_index()
```

```python
label=df['number of bedrooms'].unique()
label.sort()
label
```

```python

plt.figure(figsize=(10,7))
plt.pie(bedroom,labels= label, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

# Bivariate Analysis

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x='number of bedrooms', y='living area', data=df)
plt.title('Scatter Plot')
plt.xlabel('number of bedrooms')
plt.ylabel('living area')
plt.show()
```

# Multivariate Analysis

```python
df.corr()
```

```python
plt.figure(figsize=(15,15))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
